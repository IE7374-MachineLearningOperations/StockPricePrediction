```python
import tensorflow_data_validation as tfdv
import pandas as pd

def validate_financial_data(df_cleaned):
    # Reset the index and create a fresh DataFrame within the function
    df_new = df_cleaned.reset_index(drop=True)
    
    # Generate statistics from the DataFrame
    stats = tfdv.generate_statistics_from_dataframe(df_new)
    
    # Infer schema based on generated statistics
    schema = tfdv.infer_schema(stats)

    # Customize schema to fit specific rules
    for feature in schema.feature:
        # Rule 1: Allow specific features to have negative values
        if feature.name in ['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA', 'T10Y3M', 'DCOILWTICO', 'ads_index']:
            # No min or max constraints; we skip adding constraints here
            continue
        else:

            pass

        # Rule 2: Presence validation for required features
        if feature.name in ['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA', 'RF', 'DGS10']:
            # Set required fields with no null values allowed
            feature.presence.min_fraction = 1.0
        else:
            # Optional fields, null values allowed
            feature.presence.min_fraction = 0.0
    
    # Detect anomalies by validating stats against the schema
    anomalies = tfdv.validate_statistics(stats, schema)
    
    # Additional check: flag any columns with negative values for non-allowed features
    non_negative_features = [f for f in df_new.columns if f not in ['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA', 'T10Y3M', 'DCOILWTICO', 'ads_index']]
    negative_columns = [col for col in non_negative_features if (df_new[col] < 0).any()]

    if negative_columns:
        print("Warning: Found negative values in non-negative columns.")
        print("Columns with anomalies:", negative_columns)
    else:
        print("No negative values found in restricted columns.")
    
    # Display any schema anomalies found
    if anomalies.anomaly_info:
        print("Anomalies found in data:")
        for feature_name, anomaly_info in anomalies.anomaly_info.items():
            print(f"{feature_name}: {anomaly_info}")
    else:
        print("No anomalies found in schema validation. Data quality is good.")
    
    return anomalies, negative_columns

# Example usage
# Ensure df_cleaned is defined and valid before calling the function
anomalies, negative_columns = validate_financial_data(df_cleaned)
```
